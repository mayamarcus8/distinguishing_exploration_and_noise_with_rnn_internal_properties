```python
#config num of agent to simulate
num_of_agents = 100

#config num of trails for each block
num_of_trials = 200

#config num of block
num_of_block = 3

# for cross valdation 
array = np.arange(num_of_block)
cv = [np.roll(array,i) for i in range(num_of_block)]
cv = np.array(cv)

models = {
    'mf':[],
    'mb':[],
    'habit':[],
    'wsls':[],
    'kdh':[],
    'hybrid':[]
}
```

```python
all_data = [] 
for sim in range(num_of_block):
    data_per_block = []
    for m in tqdm(models): 
        for agent in range(num_of_agents):
            data_per_block.append((pd.read_csv(f'../data/{m}/{m}_agent_{agent}_sim_{sim}.csv')))
    all_data.append(data_per_block)
    
block_0 = all_data[0]
block_1 = all_data[1]
block_2 = all_data[2]
```

```python
N = len(block_0)

ll_lr_in_sample = np.zeros(shape=(N,3)) 
ll_lr_out_sample = np.zeros(shape=(N,6)) 
betas_0 = []

for subj in tqdm(range(N)):
    data = [block_0[subj],block_1[subj], block_2[subj]]
    
    for block in range(3):
        train = data[block%3]
        test_0 = data[(block+1)%3]
        test_1 = data[(block+2)%3]
        
        X_tr, y_tr = preprocess_logistic_regression(train)
        X_te_0, y_te_0 = preprocess_logistic_regression(test_0)
        X_te_1, y_te_1 = preprocess_logistic_regression(test_1)
    
        m,inter,coef = fit_logistic_regression(X_tr,y_tr)
        betas_0.append((m,inter,coef))
        
        if m == None:
            print('hi')
            ll_lr_in_sample[subj,block] = 0 
            
            ll_lr_out_sample[subj,block*2] = 0
            ll_lr_out_sample[subj,(block*2)+1] = 0
        else:
            ll_lr_in_sample[subj,block] = -nlp_logistic_regression(m,X_tr,y_tr)
            
            ll_lr_out_sample[subj,block*2] = -nlp_logistic_regression(m,X_te_0,y_te_0)
            ll_lr_out_sample[subj,(block*2)+1] = -nlp_logistic_regression(m,X_te_1,y_te_1)
            
ll_lr_out_sample = ll_lr_out_sample.mean(axis=1)
```

```python
res_0 = []

for _ in range(5):
    pool = mp.Pool(processes=mp.cpu_count())
    fit = pool.map(hybrid3_fit,block_0)
    pool.close()
    res_0.append(fit)

res_1 = []

for _ in range(5):
    pool = mp.Pool(processes=mp.cpu_count())
    fit = pool.map(hybrid3_fit,block_1)
    pool.close()
    res_1.append(fit)

res_2 = []

for _ in range(5):
    pool = mp.Pool(processes=mp.cpu_count())
    fit = pool.map(hybrid3_fit,block_2)
    pool.close()
    res_2.append(fit)
```

```python
N = len(block_0)
all_ll_0 = np.zeros(shape=(5,N))
for i in range(5):
    all_ll_0[i] = np.array([res_0[i][j].fun for j in range(N)])
    
all_ll_1 = np.zeros(shape=(5,N))
for i in range(5):
    all_ll_1[i] = np.array([res_1[i][j].fun for j in range(N)])
    
all_ll_2 = np.zeros(shape=(5,N))
for i in range(5):
    all_ll_2[i] = np.array([res_2[i][j].fun for j in range(N)])
    
ind_0 = np.argmin(all_ll_0,axis=0)
ind_1 = np.argmin(all_ll_1,axis=0)
ind_2 = np.argmin(all_ll_2,axis=0)

fit_t_0 = []
fit_t_1 = []
fit_t_2 = []

for i,ind in enumerate(ind_0):
    fit_t_0.append(res_0[ind][i])
    
for i,ind in enumerate(ind_1):
    fit_t_1.append(res_1[ind][i])
    
for i,ind in enumerate(ind_2):
    fit_t_2.append(res_2[ind][i])
    
def bce(y_hat,y_true):
    eps = 1e-7
    return -np.sum( y_true*np.log(y_hat+eps) + (1-y_true)*np.log(1-y_hat+eps))

LL_out = np.zeros(shape=(N,6)) 

for i in range(N):
    _, y_hat, _ = hybrid3_predict(block_1[i],fit_t_0[i].x)
    LL_out[i,0] = -bce(1-y_hat, block_1[i]['action_stage_1'].values)
    
    _, y_hat, _ = hybrid3_predict(block_2[i],fit_t_0[i].x)
    LL_out[i,1] = -bce(1-y_hat, block_2[i]['action_stage_1'].values)
    
    _, y_hat, _ = hybrid3_predict(block_2[i],fit_t_1[i].x)
    LL_out[i,2] = -bce(1-y_hat, block_2[i]['action_stage_1'].values)
    
    _, y_hat, _ = hybrid3_predict(block_0[i],fit_t_1[i].x)
    LL_out[i,3] = -bce(1-y_hat, block_0[i]['action_stage_1'].values)
    
    _, y_hat, _ = hybrid3_predict(block_0[i],fit_t_2[i].x)
    LL_out[i,4] = -bce(1-y_hat, block_0[i]['action_stage_1'].values)
    
    _, y_hat, _ = hybrid3_predict(block_1[i],fit_t_2[i].x)
    LL_out[i,5] = -bce(1-y_hat, block_1[i]['action_stage_1'].values)

LL_hybrid3_all = LL_out.mean(axis=1)
```

```python
df = pd.read_csv('full_data.csv')

models = {
        'habit':['$\it{habit}$',sns.color_palette("tab10")[2],'Habit'],
        'kdh':['$\it{kDH}$',sns.color_palette("tab10" )[4],'k-Dominant Hand'],
        'hybrid3':['$\it{Hybrid}$',sns.color_palette("tab10" )[9],'Hybrid']
}


from scipy import stats

fig = plt.figure(figsize=(14, 5))
gs = fig.add_gridspec(1,3, hspace=0, wspace=0.1 )
axes = gs.subplots(sharey=True)

for i,m in enumerate(models):

    cor = (df['opt_ep'][i*100:i*100+100]).values
    
    data = df[df['model']==m]
    cor = (data['opt_ep']).values
    data = df[df['model'] == m]
    data.reset_index(inplace=True)
    sns.scatterplot(ax=axes[i],y=1-data['noise'],x=cor, 
                    marker="$\circ$", ec="face", s=60 , color=models[m][1])
    sns.regplot(ax=axes[i],y=1-data['noise'],x=cor,scatter=False,color='k')
#     axes[i].plot([0,1000],[0,1],'--k')
    axes[i].grid(alpha=.3)
    axes[i].set_title(models[m][2],size=18)
    a,b = stats.pearsonr(y=1-data['noise'],x=cor)
    print(b)
#     axes[i].set_yticklabels(labels=[0,.2,.4,.6,.8,1],size=14)
    axes[i].text(x=10,y=.95, s='${\u03C1}$=%.2f'%a,size=18)
    


axes[0].set_ylabel('${1 \ / \ Noise}$',size=18)
axes[1].set_xlabel('Epoch',size=18)


axes[2].annotate("Agent",
                xy=(414, 0.255), xycoords='data',
                xytext=(513, 0.15), textcoords='data',
                arrowprops=dict(arrowstyle="->", color="k",connectionstyle="arc3,rad=0.3"),
                )

plt.savefig('../plots/fig_2',dpi=300)


plt.show()
```
